```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

df_kaggle = pd.read_csv("../data/kaggle-uber-other-federal.csv")
df_kaggle.head()

```

```python
fig, ax = plt.subplots()
status = sorted(df_kaggle['Status'].unique())
tab=pd.crosstab(index=df_kaggle['Status'], columns='count')

x = status
y = tab['count']

ax.bar(x, y)
plt.show()

```
