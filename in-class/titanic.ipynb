```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

df_titanic = pd.read_csv("../data/titanic.csv")
df_titanic.head()
```

```python
fig, ax = plt.subplots()
status = sorted(df_titanic['Survived'].unique())
tab=pd.crosstab(index=df_titanic['Survived'], columns='count')

x = status
y = tab['count']

ax.bar(x, y)
ax.set_xticks(x)
ax.set_xticklabels(['Dead', 'Survived'])
ax.set_ylabel('Number of Lives')
plt.show()

# plt.savefig("titanic.pdf", bbox_inches='tight', pad_inches=2)
```
